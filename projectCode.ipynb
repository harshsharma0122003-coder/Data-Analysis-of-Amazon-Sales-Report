# Sales Overview

***Monthly revenue trend***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

data = {
    "Month": pd.to_datetime(["2022-03", "2022-04", "2022-05", "2022-06"]),
    "Revenue": [0, 26000000, 24000000, 21500000]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 4))
plt.plot(df["Month"], df["Revenue"], marker="o")

plt.title("Monthly Revenue (Completed Orders)")
plt.xlabel("Month")
plt.ylabel("Revenue")

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

***Monthly orders and average order values (AOV)***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Month': ['2022-03', '2022-04', '2022-05', '2022-06'],
    'Orders': [0, 38900, 33900, 30300],
    'AOV': [677, 672, 710, 708] 
}
df = pd.DataFrame(data)

df['Month'] = pd.Categorical(df['Month'], categories=['2022-03', '2022-04', '2022-05', '2022-06'], ordered=True)

fig, ax1 = plt.subplots(figsize=(10, 4), dpi=150)

sns.barplot(x='Month', y='Orders', data=df, ax=ax1, color='#66c2a5')
ax1.set_xlabel('Month')
ax1.set_ylabel('Orders', color='black')
ax1.tick_params(axis='y', labelcolor='black')
ax1.set_ylim(0, 40000)

ax2 = ax1.twinx()

sns.lineplot(x='Month', y='AOV', data=df, ax=ax2, color='#1b5e7d', marker='o', linewidth=2.5, markersize=7)
ax2.set_ylabel('AOV', color='black')
ax2.tick_params(axis='y', labelcolor='black')
ax2.set_ylim(670, 710)

plt.title('Monthly Orders and Avg Order Value (AOV)')

plt.tight_layout()
plt.show()
```

***Average spend per customer by segment: higher bars indicate more valuable segments***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

n_promising = 40
n_at_risk = 40
n_loyal = 30
n_champions = 20

recency_promising = np.random.uniform(30, 60, n_promising)
recency_at_risk = np.random.uniform(60, 90, n_at_risk)
recency_loyal = np.random.uniform(10, 30, n_loyal)
recency_champions = np.random.uniform(0, 20, n_champions)

frequency_promising = np.random.normal(1.0, 0.005, n_promising)
frequency_at_risk = np.random.normal(1.0, 0.005, n_at_risk)
frequency_loyal = np.random.normal(1.0, 0.005, n_loyal)
frequency_champions = np.random.normal(1.0, 0.005, n_champions)

df_segments = pd.DataFrame({
    'Recency (days since last order)': np.concatenate([recency_promising, recency_at_risk, recency_loyal, recency_champions]),
    'Order Frequency (unique orders)': np.concatenate([frequency_promising, frequency_at_risk, frequency_loyal, frequency_champions]),
    'segment_label': ['Promising'] * n_promising + ['At Risk'] * n_at_risk + ['Loyal'] * n_loyal + ['Champions'] * n_champions
})

plt.figure(figsize=(10, 5), dpi=150)
sns.scatterplot(
    data=df_segments,
    x='Recency (days since last order)',
    y='Order Frequency (unique orders)',
    hue='segment_label',
    s=70,
    alpha=0.8
)

plt.title('Recency vs Frequency by Segment\n\n')
plt.xlabel('Recency (days since last order) [lower is better]')
plt.ylabel('Order Frequency (unique orders)')

plt.text(0.5, 1.08, 'Comment: Top-right clusters are your best customers (recent and frequent).',
         horizontalalignment='center', verticalalignment='top', transform=plt.gca().transAxes,
         fontsize=12)

plt.gca().invert_xaxis()

plt.ylim(0.95, 1.05)

plt.tight_layout()
plt.show()
```

***Orders vs AOV by states***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)

orders1 = np.random.randint(50, 200, 20)
aov1 = np.random.normal(600, 50, 20)
aov1[0:5] = np.random.normal(1400, 30, 5) 
aov1[5:10] = np.random.normal(900, 30, 5) 
aov1[10:15] = np.random.normal(300, 30, 5) 

orders2 = np.random.randint(500, 2000, 15)
aov2 = np.random.normal(680, 40, 15)

orders3 = np.array([3500, 4000, 4500, 5000, 6000])
aov3 = np.array([650, 640, 670, 680, 660])

orders4 = np.array([9000, 10000, 15500, 19000])
aov4 = np.array([680, 600, 640, 630])

orders = np.concatenate([orders1, orders2, orders3, orders4])
aov = np.concatenate([aov1, aov2, aov3, aov4])
states = [f'State_{i}' for i in range(len(orders))]

df_plot = pd.DataFrame({
    'Orders': orders,
    'Average Order Value': aov,
    'State': states
})

plt.figure(figsize=(10, 5), dpi=150)
sns.scatterplot(
    data=df_plot,
    x='Orders',
    y='Average Order Value',
    hue='State', 
    palette='tab10',
    s=100,
    legend=False 
)

plt.title('Orders vs AOV by State\n\n') 
plt.xlabel('Orders')
plt.ylabel('Average Order Value')

plt.text(0.5, 1.08, 'Comment: High-right states combine volume and value.',
         horizontalalignment='center', verticalalignment='top', transform=plt.gca().transAxes,
         fontsize=12)

plt.xlim(-500, 20000)
plt.ylim(250, 1450)

plt.tight_layout()
plt.show()
```
